You will be using the **Human Activity Recognition (HAR) dataset** which is a collection of data aimed at identifying human activities based on measurements from smartphone sensors. Specifically, it contains data from accelerometers and gyroscopes of smartphones carried by subjects performing activities like walking, sitting, standing, lying down, walking upstairs, and walking downstairs. This dataset is widely used for research in areas such as activity recognition, human-computer interaction, and ubiquitous computing. It provides a valuable resource for developing and testing algorithms that can automatically detect and classify human activities based on sensor data.


Run the next cell to load the dataset

```python
# Code for downloading the dataset and loading it
import pandas as pd
import numpy as np
import requests
import zipfile
import io

# URL for the dataset archive
HAR_DATASET_URL = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip'

# Download the dataset
print("Downloading the dataset...")
response = requests.get(HAR_DATASET_URL)
z = zipfile.ZipFile(io.BytesIO(response.content))
print("Download complete. Extracting files...")

# Extract the dataset
z.extractall()

# Define file paths for training and test sets
TRAIN_DATA_PATH = 'UCI HAR Dataset/train/X_train.txt'
TRAIN_LABELS_PATH = 'UCI HAR Dataset/train/y_train.txt'
TEST_DATA_PATH = 'UCI HAR Dataset/test/X_test.txt'
TEST_LABELS_PATH = 'UCI HAR Dataset/test/y_test.txt'

# Load the dataset into pandas DataFrames
X_train = pd.read_csv(TRAIN_DATA_PATH, delim_whitespace=True, header=None)
y_train = pd.read_csv(TRAIN_LABELS_PATH, delim_whitespace=True, header=None).squeeze()

X_test = pd.read_csv(TEST_DATA_PATH, delim_whitespace=True, header=None)
y_test = pd.read_csv(TEST_LABELS_PATH, delim_whitespace=True, header=None).squeeze()

# Feature names extraction
feature_names_path = 'UCI HAR Dataset/features.txt'
features = pd.read_csv(feature_names_path, delim_whitespace=True, header=None, usecols=[1])

# Naming the feature columns
X_train.columns = features.iloc[:, 0].values
X_test.columns = features.iloc[:, 0].values

print("Data loaded successfully.")
print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)

# Optionally, load the activity labels to make the y values more readable
activity_labels_path = 'UCI HAR Dataset/activity_labels.txt'
activity_labels = pd.read_csv(activity_labels_path, delim_whitespace=True, header=None, index_col=0)
y_train.replace(activity_labels.to_dict()[1], inplace=True)
y_test.replace(activity_labels.to_dict()[1], inplace=True)
```
